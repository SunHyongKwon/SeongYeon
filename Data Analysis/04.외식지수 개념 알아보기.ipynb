### CPI 소비자 물가지수

##### 정의
- 상품과 서비스의 가격 변동을 알아보기 위해 작성하는 통계이다. <br><br>

#### 계산방법
- 품목별로 가구들이 소비한 금액을 품목별로 계산을 한다. <br><br>
- 품목별로 생활에 미치는 정도는 다 다르다. 그렇기 때문에 단순히 전체 소비자물가의 가격변동률의 평균을 곱하면 안된다. <br><br>
- 기준년도를 정해서 그 년도를 100으로 잡고 기준년도의 가격과 해당년도의 가격의 비율을 찾아서 지수를 정한다. <br><br>
- 품목별 지수에 품목별 지출비중의 가중치를 둬서 이를 품목별 지수에 곱해서 정하게 된다.<br><br>
- 가중치를 천분비를 이용해서 구하기 때문에 위에 값을 다 더한것을 100으로 나눈 것이 소비자 물가지수(CPI)가 된다. <br><br>

#### 계산식
- $W_i$ = 품목별 가중치 $P_i$ = 품목별 소비
$$ CPI = \frac {\Sigma W_i * P_i} {1000} $$



### 외식 소비자 지수
- CPI 산출 방식을 참고하고, 통계청 자료를 토대로 추측을 하는 과정이다. <br><br>
- 식당별 매출들을 다 더해서 그것을 기준년도의 매출과 비교하여서 산출을 한다. <br><br>
- 한식 , 중식 , 일식 , 양식 식당이 있을 때 한식은 일반적으로 많이 먹고, 양식 같은 경우는 특별한 날이나 한식을 외식으로 하는 비중보다는 적을 것이다. <br><br>
- 그렇기 때문에 모두 다 더해서 기준년도의 매출과 나누는 것은 정확한 외식지수를 찾는 방법이 아닐 것이라 생각된다. <br><br>
- 한식 , 중식, 일식 , 양식 매출 데이터를 찾아서 머신러닝을 통해 이들의 가중치를 찾는 것을 머신러닝을 통해 구현할 생각이다. <br><br>
- 외식지수가 쌀 소모량과 역상관관계를 가지기 때문에 외식지수를 찾으므로써 쌀 소모량 역시 예측할 수 있을 것이라 기대한다. <br><br>
