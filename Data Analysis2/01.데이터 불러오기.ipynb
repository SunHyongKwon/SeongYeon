```python
import pandas as pd
import numpy as np
```

```python
one = pd.read_csv('./Data/cost-of-living.csv')
two = pd.read_csv('./Data/cost-of-living_v2.csv')
```

```python
one.info()
```

```python
two.info()
```

#### NaN 제거하기

```python
one.dropna(how='any', inplace = True)
two.dropna(how='any', inplace = True)
```

```python
one.info()
```

```python
two.info()
```

```python
global_cost = pd.concat([one,two],ignore_index=True,axis=0)
del global_cost['Unnamed: 0']
```

### 저장하기

```python
global_cost.to_csv('./Data/global_cost.csv',index=False)
```
